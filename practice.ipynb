```python
#Create a new column Total (sum of all subjects).
df['total']=df['Math']+df['Science']+df['English']+df['Computer']
df
```

```python
#Create a new column Average.
df['average']=df['total']/4
df
```

```python
#Show students whose Average is greater than 85.
df[df['average']>=85]
```

```python
#Group by City and find average Total marks.
df.groupby('City')['average'].mean().round(2)
```

```python
#Group by Gender and find highest Math score.
df.groupby('Gender')['Math'].mean().round(2).sort_values(ascending=False)
```

```python
#Find the highest scoring student in each Class and display name
df.loc[df.groupby('Class')['total'].idxmax(), ['Class', 'Name', 'total']]
```

```python
#Create a new column Result (Pass if Average ≥ 60, else Fail).
df['Result'] = df['average'].apply(lambda x: 'Pass' if x >= 60 else 'Fail')
```

```python
#Count how many students failed.
df[df['Result'] == 'Fail'].shape[0]
```
